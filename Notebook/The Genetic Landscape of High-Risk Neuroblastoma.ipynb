# The Genetic Landscape of High-Risk Neuroblastoma

Low Median Exonic Mutation Frequency = .60/megabase (0.48 non silent)

Genes with Significant Somatic Mutation Frequencies:

    1. ALK (9.2%)
    2. PTPN11 (2.9%)
    3. ATRX (2.5%, additional 7.1% w/ focal deletions)
    4. MYCN (1.7%, recurrent p.Pro44Leu Alteration)
    5. NRAS (0.83%)

# # of Patients = 240

Patient Age = 18 months <--> 5.5 years at diagnosis with metastatic disease

Coding Regions (240 Cases)
    * 5,291 Somatic Mutations in 3,960 genes
    * Median of 18 Exomic Mutations per tumor
        *14/18 non-silent mutations 
    * Median Frequency of 0.60 mutations per megabase (.48 nonsilent/Mb)
    
Verified 241 of 282 coding candidate somatic substitutions (85%, 525 of 605 including non-coding) and 26/41 coding indels (63%, 27/79 including non-coding)

No correlation between mutation frequency and diagnosis age

Phi29-based whole genome amplification was used to generate sufficient tumor and matched normal DNA template from 80 cases. 

    *Reads were aligned to build hg19/GRCh37 of the human genome reference sequence using Burrows-Wheeler Aligner70 
    *Somatic mutations were detected using SNVmix71 (Illumina genomes and RNA-seq), muTect27 (exomes) and version 2 of the Complete Genomics’ custom caller21,72. 
    *Mutations were annotated using Oncotator and MutSig32 was used to identify genes mutated at significant frequencies. 
    *PathSeq73 was used to query exome data sets for reads supporting viral infection
