# Genomic sequencing of meningiomas identifies oncogenic SMO and AKT1 mutations

## Tissue: Meningiomas

<b>Most Frequently Mutated Gene</b> : NF2 (tumor suppressor) - disrupted in <b>50%</b> of cases

* 9 nonsense
* 9 splice-site 
* 9 frameshift indels

<b>Sequencing:</b>

    1. Discovery Set Sequencing of Grade 1 Meningiomas: 

        * WGS (57X): n = 11

        * WES (181X): n = 6


    2. Validation Set Sequencing (154X): 645 known cancer-associated genes

        * 30 additional grade 1 tumors
        * 18 genes in an extrapolation set of Grade 2/3 
            * allow for evaluation of findings for grade 1 tumors to progress to higher grades

    3. Additional 48 tumors sequenced to validate somatic genetic alterations

<b> Focal NF2 Inactivation:</b> 43% of tumors

    * Additional 8% had epigenetic modifier alterations
    * A subset of meningiomas lacking  NF2 alterations harbored recurrent oncogenic mutations in  AKT1 (p.Glu17Lys) and SMO (p.Trp535Leu) and exhibited  immunohistochemical evidence of activation of these pathways. 

<b>SCNAS:</b> loss of chr 22 most frequent (contains NF2)

* occur in all ten tumors w/ focal NF2 alteration and in 1 w/out
* Median meningioma had SCNAs affecting only 3.3% of the genome, one rearrangement and eight nonsynonymous mutations.

Among mutations, cytosine-to-thymidine transitions within a CpG context predominated, consistent with spontaneous deamination 

Rates of alteration in meningiomas were significantly lower (~10x) than prev. determined rates from other tumors

* In contrast, rates of genomic disruption were higher in the grade II and III meningiomas in the extrapolation set than for their grade I counterparts

Mean rate of nonsynonymous mutation within the 645 sequenced genes was nearly twice as high in higher-grade tumors (3.0 versus 1.6; P < 0.02), and the proportion of the genome affected by SCNAs was markedly higher (median of 12.8% versus 0.3%; P < 0.001).

<b>Candidate Somatic Rearrangements:</b> 110

* 93 confirmed reads spanning putative fusion site

* most arrangments concentrated in 3 tumors, 1 w/ chromothripsis

Also, Nonsynonymous mutations in 190 other genes


* The nonsynonymous muta- tion rates of tumors with and without focal NF2 alterations did not differ significantly, nor did the spectra of their mutation subtypes

<b>Genes of Statistical Significance:</b> 6
(1-4 prev. implicated)

    1. NF2
    2. KDM5C
    3. SMO
    4. AKT1
    5. RGPD3 
    6. CD300C

* Isolated/Rare Mut. in 23 genes (TP53, APC, CBL, STK11, NOTCH2)

Among genes mutated in our cohort and previously associated with cancer, three (KDM5C, KDM6A and SMARCB1) are epigenetic modi- fiers. Mutations involving these genes affected 8% of the cohort.

#### Mutation identified (encoding a p.Arg374Gln alteration) is near a mutational hotspot (p.Arg377His) described in meningiomas

<b>SMO Mutation</b> in 3 tumors (5%)

* None of these tumors had NF2 focal alteration

Six samples had mutations of the <b>phosphatidylinositol 3-kinase (PI3K)-AKT-mTOR pathway</b>. 

* None of these had mutations of NF2 or SMO (P = 0.03)

* Mutations in the Hedgehog and PI3K pathways were also associated with particular histopathological subtypes

Meningiomas with mutated SMO and AKT1-MTOR were predomi- nantly of the meningothelial subtype 

* contrast to the grade I NF2-altered tumors, which were predominantly fibroblastic and/or transitional

Validated these findings by genotyping an additional 46 grade I and 49 grade II or III meningiomas for AKT1 (p.Glu17Lys) and SMO (p.Trp535Leu and p.Leu412Phe) mutations <i>(using a mass spectrometry– based method (Sequenom hME))</i>.

* One SMO mutation (p.Leu412Phe)
* Three additional AKT1 mutations (p.Glu17Lys)

Observed recurrent mutations in signaling pathways and epi- genetic modifiers, but data also highlights the heterogeneity of mutations in these tumors. 

* Among the 17 tumors comprehen- sively characterized, 3 (18%) did not have mutations in any of the significantly mutated genes. 

#### The paucity of addi- tional genetic events in meningiomas harboring mutations of the PI3K and Hedgehog signaling pathways suggests that patients with these meningiomas may benefit from such targeted therapies already in development or use.
