###### 1) Validated ACC nonsynonymous mutations

## Merged Data

* Double check if Age column is correct since have to change to str so stripping whitespace does not mess it up
* Should I dropna with thresh of 2? : merged.dropna(thresh=2)
* Dropping duplicates drops one row.. I do not know why, however shouldn't be problem

```python
merged = pd.merge(patchar, nonsymmut, how='outer')
merged.drop_duplicates(inplace=True)
merged.Age.astype(str)
merged.apply(lambda x: x.str.strip())
merged.Age.astype(int)
# applied string to make pat id format consistent
merged['Patient_ID'] = merged['Patient_ID'].apply(lambda x: 'ACC' + str(x))
merged['Tissue'] = 'ACC'
merged.to_csv(MERGED_PATH+'mergedMutACC.csv')
merged.head()
```
