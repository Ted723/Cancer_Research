# The Mutational Landscape of Adenoid Cystic Carcinoma

### Patient Analysis

* 60 tumor-normal pairs w/ ACC
    * <i>(55 WES, 5 WGS)</i>

### Mutation Analysis

* <b>Low Exonic Mutation Rate</b>: 0.31 non-silent events/ Mb 
* <b> Sig Mut Genes: </b>
    * Chromatin-State Regulators
        * SMARCA2, CREBBP, KDM6A
    * Genes central to DNA Damage Response and Protein Kinase A Signaling
        * MYB-NFIB translocations
        * MYB-associated genes w/ somatic mutations
    * Recurrent Mutations in the FGF-IGF-PI3K Pathway (30% of tumors)       

### Summary

* Large-scale ACC analysis and highlight new ACC driver mutations within corresponding pathways. 
    * Convergent alterations include pathway derangements involving chromatin remodeling, DNA damage response, MYB-MYC and tyrosine kinase signaling. 
* Discovery of genomic alterations in targetable pathways suggests potential avenues for novel treatments to address a typically chemoresistant malignancy. 
    * Verified ACC cell lines are needed to further substantiate the clinical usefulness of the mutations identified here. 
* In total, data provides insights into the genetic framework underlying ACC oncogenesis and establish a foundation for identifying new therapeutic strategies.
