# Exome Sequencing Identifies Distinct Mutational Patterns in Liver Fluke-Related and Non-Infection-Related Bile Duct Cancers

### Patient Analysis

* 209 Patients with Cholangiocarcinomas (CCAs)
    * WGS/WES Analysis
    * from Asia and Europe
    * 108 Infection w/ liver fluke Opisthorchis Viverrini
    * 101 Non O.viverrini-related etiologies
    * Median Survival Time => 24 months

### Mutational Analysis

* 245 Somatic Mutations in 224 Genes 
    1. 193 Missense (78.8%)
    2. 21 Indels (8.6%)
    3. 20 Nonsense (8.2%)
    4. 11 Splice Site (4.5%)
    
* Non-O.viverrini = Avg. 16 Somatic mutations/tumor
* O.viverrini = Avg. 26 Mutations/tumor
* <b>New recurrently mutated genes </b> (7) in CCA, including: AGPAT6, ATP10A, BRPF3, CCT8L2, GPR112, HMCN1 and LRRIQ1
* Recurrent Somatic Mutations in CCAs (WGS)
    * BAP1: 11 indels, 3 nonsense, 2 missense
    * ARID1A: 17 indels, 14 nonsense, 3 missense, 1 splice
        * Both demonstrated tumor suppressive functions
            * Establishes role of chromatin modulators in CCA pathogenesis

    * BAP1, IDH1 and IDH2 - more frequently mutated in non–O. viverrini CCAs
    * TP53 mutations showed the reciprocal pattern
* These findings indicate that different causative etiologies may induce distinct somatic alterations, even within the same tumor type.
* Suggests non-liver fluke related risk factos contribute lower mutagenic potency

### Summary

* Studies suggest that different carcinogenic etiologies may induce distinct mutational landscapes in CCA, even within the same tumor type.
* Furthermore, non-liver fluke related risk factors contribute lower mutagenic potency
