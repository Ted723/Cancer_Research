# The Genetic Landscape of High-Risk Neuroblastoma

### Patient Analysis

* 240 w/ High-Risk Neuroblastoma
    * WGS, WES, and Transcriptome Analysis
    * Part of TARGET initiative
* Age = 18 months <--> 5.5 years at diagnosis w/ metastic disease

### Mutation Analysis

* Low Median Exonic Mutation Frequency = .60/megabase (0.48 non silent)

* FMGs: ALK, PTPN11, ATRX, MYCN

* Coding Regions (240 Cases)
    * 5,291 Somatic Mutations in 3,960 genes
    * Median of 18 Exomic Mutations per tumor
        *14/18 non-silent mutations 
    * Median Frequency of 0.60 mutations per megabase (.48 nonsilent/Mb)
    
* Verified 241 of 282 coding candidate somatic substitutions (85%, 525 of 605 including non-coding) and 26/41 coding indels (63%, 27/79 including non-coding)

### Summary

* Genes with significant somatic mutation frequencies included ALK (9.2% of cases), PTPN11 (2.9%), ATRX (2.5%, and an additional 7.1% had focal deletions), MYCN (1.7%, causing a recurrent p.Pro44Leu alteration) and NRAS (0.83%). 
* Rare, potentially pathogenic germline variants were significantly enriched in ALK, CHEK2, PINK1 and BARD1. 
* The <b>relative paucity of recurrent somatic mutations</b> in neuroblastoma challenges current therapeutic strategies that rely on frequently altered oncogenic drivers.

* No correlation between mutation frequency and diagnosis age

* Phi29-based whole genome amplification was used to generate sufficient tumor and matched normal DNA template from 80 cases. 
    * Reads were aligned to build hg19/GRCh37 of the human genome reference sequence using Burrows-Wheeler Aligner70 
    * Somatic mutations were detected using SNVmix71 (Illumina genomes and RNA-seq), muTect27 (exomes) and version 2 of the Complete Genomics’ custom caller21,72. 
    * Mutations were annotated using Oncotator and MutSig32 was used to identify genes mutated at significant frequencies. 
    * PathSeq73 was used to query exome data sets for reads supporting viral infection

```python
patchar = pd.read_table(PATH + 'patchar.csv', error_bad_lines=False, sep = ',', header=1)
patchars = patchar[['Case USI','TARGET ID','Analysis ID','Data source (Exome/Genome)',
                    'AGE','GENDER','NAACCR_ETHNICITY','NAACCR_RACE','year','hist',
                    'grade','efstime','stime']]
patchars['AGE'] =patchars['AGE'].apply(lambda x: x/12/30.)
patchars.rename(columns={'Case USI':'Patient_ID','AGE':'Age',
                         'GENDER':'Gender'}, inplace=True)
patchars.to_csv(MUT_PATH+'patcharHRNeuro.csv')
patchars.head()
```

### Mutation Data

```python
a = pd.read_csv(PATH + 'mutdata.csv')
a.drop(['MUTSIG_Published_Results','OREGANNO_ID','FamilialCancerDatabase_Syndromes',
        'DNARepairGenes_Role','CGC_Tumor_Types_Germline',
        'CGC_Other_Diseases','CGC_Tumor_Types_Germline'],1, inplace=True)
mutdata = a[['Hugo_Symbol','Chromosome','Variant_Classification',
             'Variant_Type','Genome_Change','Protein_Change','Case_USI',
             'Entrez_Gene_Id','Data_Type','Tumor_Sample_Barcode']]
mutdata.rename(columns = {'Case_USI':'Patient_ID','Hugo_Symbol':'Gene',
                          'Chromosome':'Chr','Variant_Classification':'Mutation_Category',
                          'Variant_Type':'Mutation_Type'}, inplace=True)
mutdata.to_csv(MUT_PATH+'mutdataHRNeuro.csv')
mutdata.head()
```

### Merged Data

```python
merged = pd.merge(patchars, mutdata, how='right')
# following two lines make gender format consistent
merged.Gender = merged.Gender.astype(str)
merged.Gender.replace(['1','2'],['M','F'], inplace=True)
merged.to_csv(MUT_PATH+'mergedHRNeuro.csv')
merged.head()
```
