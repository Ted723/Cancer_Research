# Exome and whole-genome sequencing of esophageal adenocarcinoma identifies recurrent driver events and mutational complexity 

### Patient Analysis

* 149 Esophageal Adenocarcinoma (EAC) tumor-normal pairs WES 
* 15 EAC tumor-normal pairs WGS
    * Originally 16, but one was thrown out 
* Chromosome Build: 37 - hg19

### Mutation Analysis

* <b>Mutational Signature:</b> high prevalence of A>C transversions at AA dinucleotides
* <b>Significantly Mutated Genes:</b>  26
    * <i>Previously Implicated (old)</i> - TP53, CDKN2A, SMAD4, ARID1A and PIK3CA
    * <i>New</i> - SPG20, TLR4, ELMO1 and DOCK2 (chromatin-modifying factors and candidate contributors)

### Summary

* <b>ELMO1:</b> EAC derived mutations identifies increased cellular invasian
* <b>Belief:</b>
    * Activation of RAC1 pathway as a contributor to EAC tumorgenesis
        * ELMO1 and DOCK2 both upstream modulators of the RAC1 GTPase
* <b>Candidate Rearrangements:</b> 2,952 were identified
    * Median of 172 per tumor (range of 77–402)
    * Median of 26,161 mutations across the genome per tumor (range of 18,881–66,225)
    * Corresponding to a median mutation frequency of 9.9 mutations/Mb (range of 7.1–25.2 mutations/Mb) relative to a hap-loid genome. 
    * The median mutation frequency was highest in inter-genic regions (13.9 mutations/Mb), intermediate in intronic regions(8.7 mutations/Mb) and lowest in coding exons (6.6 mutations/Mb)


### Accession Codes

<b>Accession codes:</b>
Binary sequence alignment/map (BAM) files were deposited in the database of Genotypes and Phenotypes (dbGaP) under accession <b>phs000598.v1.p1</b>. Raw mRNA expression data for 14 EAC samples have been deposited at the Gene Expression Omnibus (GEO) under accession <b>GSE42363</b>.

### Hard Data

#### Patient Data

##### Sup Table 1.  EAC Sample Cohort for WGS and WES

```python
cohort = pd.read_table(PATH + 'tabEACsamplecohort.csv', usecols=['Sample ID','WGS', 'WES','Gender','Age', 'Location','Ethnicity','Tumor Grade','Condensed Location'])
cohort.rename(columns={'Sample ID':'Patient_ID','Tumor Grade':'Tumor_Grade'}, inplace=True)
cohort.to_csv(MUT_PATH+'patcharExEsophAdeno.csv')
cohort.head()
```

### Mutation Data

#### Supplementary Table 14. Somatic mutations in MSI positve samples detected by WES (n=4).

```python
df = pd.read_csv(PATH + 'ng.2591_S5.csv', usecols=['Genome Change','Sample Name','Gene',
                                            'Chromosome','Mutation Type','Mutation Classification',
                                            'Protein Change'])
a = df.rename(columns={'Genome Change':'Genome_Change','Chromosome':'Chr',
                       'Mutation Type' : 'Mutation_Type','Mutation Classification':'Mutation_Category',
                       'Protein Change':'Protein_Change'})
#took out tumor part of string in order to keep Patient_ID format consistent and allow merge
a['Patient_ID'] = a['Sample Name'].str.replace('-Tumor', '')
a.to_csv(MUT_PATH + 'mutdataExEsophAdenoMSI+.csv')
a.head()
```

###### Supplementary Table 15. Somatic mutations in non-MSI positve samples detected by WES (n=145).

* Concerns: 
    * Some genes are numbered

```python
x = pd.read_csv(PATH + 'ng.2591_S6.csv')
y = x[['Genome Change','Sample Name','Gene','Chromosome','Mutation Type','Mutation Classification','Protein Change']]
z = y.rename(columns={'Genome Change':'Genome_Change','Chromosome':'Chr','Mutation Type' : 'Mutation_Type','Mutation Classification':'Mutation_Category','Protein Change':'Protein_Change'})
#same as above.. making Patient_ID format consistent
z['Patient_ID'] = z['Sample Name'].str.replace('-Tumor','')
z.to_csv(MUT_PATH+'mutdataExEsophAdeno-nonMSI.csv')
z.head()
```

### Merged Data

```python
merged = pd.merge(a, z, how='outer')
merged1 = pd.merge(merged, cohort, how='outer')
#strip whitespace
merged1.apply(lambda x: x.str.strip())
merged1.drop('Sample Name', 1, inplace=True)
merged1['Tissue'] = 'Esophageal Adenocarcinoma'
merged1.to_csv(MERGED_PATH + 'mergedExEsophAdeno.csv')
merged1.head()
```
