### Read in the Files

```python
#reads in the csv files
def reader(f):
    d = read_csv(f, header=0,  sep = ',')
    ## do not know why i put reshape in so I've commented it out
    #d.columns = range(d.shape[1])
    return d

# here can specify keys and whether or not to ignore index
df = concat([reader(f) for f in files], keys=files, ignore_index=False)

#show full dataframe of values
df

#slice out parts of dataframe for primary analysis
a = df[['Patient_ID','Age','Chr','Gender','Gene', 'Protein_Change','Tissue','Genome_Change','Mutation_Type','Mutation_Classification']]

##show dataframe of interest
a

### Optional tasks: ###

##drop level of index so that there is no integer index.. can drop file index by doing level = 0

a.reset_index(level=1, drop=True, inplace=True)
a

##One way of querying and setting up hierarchical indexes
#b= a.query('Gender == "F"')[['Gene','Protein_Change']]
#b
```

#### Clean up DataFrame

take out chr in strings in chr column to make consistent...

```python
a.Age = a.Age.astype(str)
a = a.apply(lambda x : x.str.strip())
a.Chr = a.Chr.str.replace('chr', '')
```

Unify gender across sudies

```python
a.Gender.replace(['Male','Female','1','2'],['M','F','M','F'], inplace=True)
a.Age = a.Age.astype(float64)
```

#### Clean up mutation naming

* Questions???:can i change p.235Leu>Pro to  p.L235P ? assuming thats the right notation 
* might need to remove ending asterisks in some genes

```python
a.Mutation_Type.replace([('Missense_Mutation','missense','Missesne')],('Missense'), inplace=True)
a.Mutation_Type.replace([('Nonsense_Mutation','nonsense')],('Nonsense'), inplace=True)
a.Mutation_Type.replace([('Nonstop_Mutation','stop codon read through')],('Nonstop'), inplace=True)
a.Mutation_Type.replace([('utr-3')],('3\'UTR'), inplace=True)
a.Mutation_Type.replace([('utr-5','5\'Flank')],('5\'UTR'), inplace=True)
a.Mutation_Type.replace([('frameshift deletion')],('Frame_Shift_Del'), inplace=True)
a.Mutation_Type.replace([('frameshift insertion')],('Frame_Shift_Ins'), inplace=True)
a.Mutation_Type.replace([('inframe insertion')],('In_Frame_Ins'), inplace=True)
a.Mutation_Type.replace([('inframe deletion')],('In_Frame_Del'), inplace=True)
a.Mutation_Type.replace([('Splice Site','splice','splicesite','Splice site','spliceSite')],('Splice_Site'), inplace=True)
a.Mutation_Type.replace([('indel')],('Indel'), inplace=True)
a.Mutation_Type.replace([('Frame-shift_indel')],('Frame_Shift_Indel'), inplace=True)
a.Mutation_Type.replace([('synonymous-coding')],('Synonymous-Coding'), inplace=True)

a.Mutation_Classification.replace([('Missense')],('Missense_Mutation'), inplace=True)
```
