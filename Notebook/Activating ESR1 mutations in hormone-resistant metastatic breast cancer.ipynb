# Activating ESR1 mutations in hormone-resistant metastatic breast cancer

### Patient Analysis

* 11 w/ ER+ metastatic breast cancer
    * 6/11 patients w/ ESR1 mutation
    * WES and Transcriptome Analysis
    * 3 of cases primary diagnostics showed ESR1 mutations were not present at an earlier stage, indicating that they were aquired after endocrine therapy

### Mutation Analysis

* 39 gene fusions in the 6 index patients
    * 11 encoding in frame fusion proteins

### Sumarry

* 11 Patients w/ ER+ metastatic breast cancer
* 6/11 patients:  mutations in ESR1 
* The most notable observation in the mutational landscapes of these treated patients with ER-positive breast cancer was the finding of nonsynonymous mutations in ESR1 affecting the LBD (n = 6). 
* The six index patients MO_1031, MO_1051, MO_1069, MO_1129, MO_1167 and MO_1185 had mutations encoding p.Leu536Gln, p.Tyr537Ser, p.Asp538Gly, p.Tyr537Ser, p.Asp538Gly and p.Tyr537Ser alterations in the LBD, respectively
 affects ligand binding domain (5 LBD alterations identified)
 * all had been prev. treated w/ anti-estrogens and estrogen deprivation therapies
 * Each occurred in the vicinity of the synthetic alterations of ESR1 that are inverted in response to tamoxifen and involve p.Met543Ala and p.Leu544Ala alterations (Inv-mut-AA2)15 and served as a positive control for our subsequent in vitro studies !! !possps

 * After looking at TCGA, Here we took advantage of data from the TCGA Project, which has generated whole-exome sequences for 27 tumor types across at least 4,000 individual samples. As expected, LBD-disrupting mutations of ESR1 were not detected in the 390 ER-positive breast cancers sequenced by TCGA, as these were primary resection samples before hormonal treatment16, nor did we detect ESR1 mutations in a cohort of 80 triple-negative breast carcinoma transcriptomes  !!!
    
* *Suggests: activating mutations in ESR1 are a key mechanism in acquired endocrine resistance in breast cancer therapy.

### Mutational Analysis

5 LBD-localized ESR1 non-synonymous mutations

* p.Leu536Gln, p.Tyr537Ser, p.Tyr537Cys, p.Tyr537Asn and p.Asp538Gly

    * resulted in constitutive activity and continued responsiveness to anti-estrogen therapies <b> in vitro
    * Each occurred in the vicinity of the synthetic
alterations of ESR1 that are inverted in response to tamoxifen and
involve p.Met543Ala and p.Leu544Ala alterations (Inv-mut-AA2) 15
and served as a positive control for our subsequent in vitro studies.
