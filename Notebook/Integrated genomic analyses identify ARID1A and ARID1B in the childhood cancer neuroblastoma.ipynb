## Hard Data

### Patient Data:

* Converted age in days to years, since child... may need to change in the future
* Concerns: no histology, gender, race
* When reading in csv, started header on 2 to format better

```python
patdata = pd.read_csv(PATH + 'sampledata.csv', header=2, 
                      usecols=['Whole-Genome Sequencing for Mutational Analysis',
                               'Whole-Exome Capture Sequencing for Mutational Analysis',
                               'Sample Name','Age at Diagnosis (Days)','EFS Time (Days)',
                               'S Time (Days)',
                               'Genome-Wide Copy Number by Paired-End Tag Analysis'])
patdata.rename(columns={'Age at Diagnosis (Days)':'Age','Sample Name':'Patient_ID',
                        'Whole-Genome Sequencing for Mutational Analysis':'WGS',
                        'Whole-Exome Capture Sequencing for Mutational Analysis':'WES'
                        },inplace=True)
#changed age to account for fact it is more specific since they are children
patdata.Age = patdata.Age.apply(lambda x: x/30/12)
patdata.to_csv(MUT_PATH + 'patcharARID1Achildneuro.csv')
patdata
```

### Mutation Data:

```python
sommut = pd.read_csv(PATH + 'sommut.csv', header=1, skipfooter=1, 
                     usecols=['Sample','Nucleotide (genomic)','Gene Symbol',
                              'Amino Acid (protein)','Mutation Type'])
sommut.rename(columns={'Sample':'Patient_ID','Gene Symbol':'Gene',
                       'Nucleotide (genomic)':'Genome_Change',
                       'Amino Acid (protein)':'Protein_Change',
                       'Mutation Type':'Mutation_Type'}, inplace=True)
#added g. to make format consistent
sommut.Genome_Change = sommut.Genome_Change.apply(lambda x: 'g.' + str(x))
sommut.to_csv(MUT_PATH + 'mutdataIntgaARID1A.csv')
sommut
```

### Merged Data

* Not everything has values due to missing data

```python
merged = pd.merge(patdata, sommut, how='right')
merged.dropna(thresh=4, axis=0, inplace=True)
merged['Tissue'] = 'Childhood Cancer Neuroblastoma'
merged.to_csv(MERGED_PATH + 'mergedARID1Achildneuro.csv')
merged
```
