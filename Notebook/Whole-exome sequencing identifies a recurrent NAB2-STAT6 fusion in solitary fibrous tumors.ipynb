* Avg. # of Non-Synonymous Mutations: 22.5
* 390 Somatically Mutated Genes

### Summary

* The NAB2-STAT6 fusion seems to be unique to SFT samples
    * Fusion analysis of approximately 713 unique tumor-normal pairs from 5 tumor types analyzed by whole-genome, whole-exome or whole-transcriptome sequencing or a combination of these techniques did not identify any fusions involving these genes
*  Analysis in 53 tumors confirmed the presence of 7 variants of this fusion transcript in 29 tumors (55%)
    * Represents a lower bound for fusion frequency at this locus
        * Suggests that the NAB2-STAT6 fusion is a distinct molecular feature of SFTs.
